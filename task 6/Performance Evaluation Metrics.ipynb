```python
import pandas as pd

df = pd.read_csv('intern_performance_1.csv', parse_dates=['assigned_date', 'submitted_date'])

# Check for missing values
print(df.isnull().sum())

# Verify completion days match date difference
df['calc_days'] = (df['submitted_date'] - df['assigned_date']).dt.days
mismatches = df[df['calc_days'] != df['task_completion_days']]
print(f"Mismatches: {len(mismatches)}")

df.to_csv('intern_performance_clean.csv', index=False)
```

```python
import pandas as pd
from datetime import datetime

df = pd.read_csv('intern_performance_clean.csv', parse_dates=['submitted_date'])

# Filter for current month
month = datetime.now().month
year = datetime.now().year
monthly = df[(df['submitted_date'].dt.month == month) & 
             (df['submitted_date'].dt.year == year)]

# Summary stats
report = monthly.groupby(['intern_name', 'department']).agg(
    tasks=('task_name','count'),
    avg_score=('overall_score','mean'),
    avg_days=('task_completion_days','mean'),
    mentor_avg=('mentor_rating','mean')
).round(2)

# Export to Excel for supervisors
report.to_excel(f'monthly_report_{year}_{month:02d}.xlsx')
print("Report saved!")
```
